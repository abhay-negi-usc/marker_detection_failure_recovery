```python
import numpy 
import json 
import os 
import matplotlib.pyplot as plt
```

```python
dir_flying_aruco = "/media/rp/Elements1/abhay_ws/FlyingArUco v2 Dataset/flyingarucov2/" 

# list all json files in directory 
json_files = os.listdir(dir_flying_aruco)
json_files = [f for f in json_files if f.endswith('.json')]

# extract data from json files 
marker_ids = [] 
for file in json_files: 
    data = json.load(open(dir_flying_aruco + file)) 
    for i in range(len(data["markers"])) :
        marker_ids.append(data["markers"][i]["id"]) 

marker_ids = numpy.array(marker_ids)
# print 10 most common marker ids 
unique_ids, counts = numpy.unique(marker_ids, return_counts=True) 
# print number of unique ids 
print("Number of unique marker ids: ", len(unique_ids)) 
# print mean number of counts 
print("Mean number of counts: ", numpy.mean(counts)) 
# order by counts 
sorted_indices = numpy.argsort(counts)[::-1]
unique_ids = unique_ids[sorted_indices]
counts = counts[sorted_indices]
print("10 most common marker ids: ")
for i in range(10): 
    print(unique_ids[i], counts[i])

# bar chart of marker ids 
plt.bar(unique_ids, counts)
plt.xlabel('Marker ID')
plt.ylabel('Count')
plt.title('Distribution of Marker IDs')
plt.xticks(unique_ids)
plt.grid(axis='y')
plt.show()    
```
